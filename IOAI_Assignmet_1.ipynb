Question 1: Create a data frame called playlist with the following information:

Eight songs of your choice in a column 'Song'. I recommend choosing your 8 favourite songs.
The artists for those songs in a column 'Artist'
The vibe for those songs in a column 'Energy'. The vibe should be on a 0 to 1 scale, with 1 being highest energy and 0 being lowest energy.
The day that the song was played in a column 'Day_Played'.
Print the playlist, then show the first 5 rows of the playlist

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

songdata = {'Favourite Songs': ['100 numbers of pi', 'Grand Finale', 'One of a Kind', '7 years', 'Bueatiful Things', '200 numbers of pi', '300 numbers of pi', 'Fart Elise'], 'Artists': ['AsapScience', 'Disnep', 'Disnep', 'Lukas Graham', 'Benson Boone', 'AsapScience', 'AsapScience', 'Fart Windbreaker'], 'Vibe Level': ['0', '1', '1', '0', '1', '0', '0', '1'], 'Day Played': ['Saturday', 'Monday', 'Monday', 'Friday', 'Tuesday', 'Saturday', 'Saturday', 'Friday']}
df = pd.DataFrame(songdata)
print(df)
```

Question 2: Evaluate which artist is your top (most listened to) artist, then print the results

```python
artistcount = df['Artists'].value_counts().idxmax()
#df['Artists'] Accesses the "Artists" column from the DataFrame (df) and retrieves the list of all the artists for each song in your dataset.
#value_counts() This counts how many times each artist appears in the "Artists" column. It aslo counts it in desending order so the most listened artist comes first.
#idxmax() This returns the index (in this case, the artist's name) of the highest value (the artist with the most songs played).
print(artistcount, 'is your most listened artist')# Prints the most listened artist
```

Question 3: Evaluate your average energy level.

If your average energy level is less than 0.4, print something interesting like "I'm just a chill guy"
If your average energy level is more than 0.6, print something interesting
If your average energy level is between 0.4 and 0.6 (inclusive of these numbers) print something interesting

```python
df['Vibe Level'] = pd.to_numeric(df['Vibe Level'])# Convert the vibe values (which are currently set as strings) to integers

avg_vibe = df['Vibe Level'].mean()# Calculate average vibe level

print(f"Average energy level: {avg_vibe:.2f}")# Print average

# Print interesting messages
if avg_vibe < 0.4:
    print("I'm just a chill guy.")
elif 0.4 <= avg_vibe <= 0.6:
    print("Balanced vibes - Just in between.")
else:
    print("High energy! I'm bringing the party wherever I go.")
```

Question 4: Create a bar chart of music played per day. Then print the following: "I think I listen to music mostly on DAYOFWEEK because..."

```python
import pandas as pd
import matplotlib.pyplot as plt

day_counts = df['Day Played'].value_counts() # Finds out how many songs were played on each day
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'] # Sets the order in which the days are displayed on the bar graph
day_counts = day_counts.reindex(days_order, fill_value=0) # For the days where you dont listen to anysongs it sets the value a 0 instead of skipping it

most_played_song = df['Favourite Songs'].value_counts().idxmax()# Find the most played song

# Find the most active listening day
most_active_day = day_counts.idxmax()
most_active_day_count = day_counts.max()


plt.bar(day_counts.index, day_counts.values, color='skyblue')# Set up and colour the graph
plt.title('Music Played Per Day')# Add title
plt.xlabel('Days')# Add the x axis label
plt.ylabel('# of Songs Played That Day')# Add the y axis label

plt.figtext(0.5, -0.05, f"Most played song: '{most_played_song}'", ha="center", fontsize=10)# Display what your most played song is
plt.figtext(0.5, -0.1, f"You listened to the most songs on: {most_active_day} ({most_active_day_count} songs)", ha="center", fontsize=10)# Display you most active day

plt.tight_layout()# Adjust spacing
plt.show()# Display the table
```
